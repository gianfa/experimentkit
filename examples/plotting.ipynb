# Plotting

```python
import matplotlib.pyplot as plt

from experimentkit.visualization import styles
```

Experimentkit comes with a custom style already enabled.
If you want to choose another custom style run `styles.keys()`, after having imported `styles` dict from *experimentkit.visualization* module.

```python
styles.keys()
```

Here an example of default style

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Custom Style Plot')
plt.show()
```

Here an example of other styles. Credit: [SciencePlots](https://github.com/garrettj403/SciencePlots).

```python
with plt.style.context('ieee'):
    plt.plot([1, 2, 3], [4, 5, 6])
    plt.xlabel('X Label')
    plt.ylabel('Y Label')
    plt.title('Custom Style Plot')
    plt.show()
```
